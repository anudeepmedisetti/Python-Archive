## Adaptation of previous example to my own Iris dataset

Please, read the following cell, I spend a lot of time trying my best to explain the following lines of code. It is good for your understanding

````python
import numpy
import matplotlib.pyplot as plt
import pandas as pd # adaptation : I like pandas library, it has useful functions like load datasets, etc.
'''
Look at the data. Open your file iris.data.txt that is stored in your computer. Then, look for :
- *The header* : is it present or not ? (header should be the first row that specifies the names of each column).

By default the function from pandas library pd.read_csv(...) considers the first row as the header, so if we
don't have a header we have to specify it like this : header=None. In the contrary, if we have a header
in the dataset we don't specify anything so we don't use this parameter.
- *The delimeters* : (please remember that each row is called an observation/example/sample/..., 
and each column is called a feature/variable/caracteristic/..., so one observation is caractarised by many
features i.e. a cat can be an observation which has a weight of 2kg and height of 50cm)
Ok so focus on one row in the file, and try to find what symbol separates each caracteristic of an observation,
is the symbol a ',' or a ';' or something else ?

By default the function from pandas library pd.read_csv considers that the default separator is a ',', so if it is
not the case in your dataset, you have to specify that to the function like this : just add the following 
argument : delimiter=';' if your separator is a ';'

- Look also for the *number of columns* i.e. for an observation like this : 5.1,3.5,1.4,0.2,Iris-setosa I counted 
4 times the delimiter ',' so it means that the dataset has 5 columns

- *The target column*. In general, the last column is often the target variable, which is the category where 
belongs the given observation i.e. 5.1,3.5,1.4,0.2,Iris-setosa When I read this row, I read the following :
The observation that has the following caracteristics : 5.1,3.5,1.4,0.2 belongs to the category Iris-setosa

Ok so focus on the values of the target column, are they integers ? string of caracters ? In order to plot
without problems our dataset in 2d with the scatter function we want integers as target values
'''
# load data using pandas
# paramater header set to None because there is no header
# remove header paramater if header is present
# add this parameter delimiter=';' if it separates each element in a row of your dataset
data = pd.read_csv('iris.data.txt', header=None)
# data is a dataframe type that belongs to the pandas library
# I can use the type(...) function to make sure of that
print(type(data)) # will print : <class 'pandas.core.frame.DataFrame'>

# So like it is an object of pandas library, I can use functions from that library.
# A useful function is head() it will diplays the first rows of the dataset, it can be useful if the 
#   dataset is large and we don't want to print a lot of things
print(data.head()) # prints the 5 first rows

# Now we will put in X the data features, and in y the target values
'''
As we saw in class, the feature values are in the 4 first columns.
So iloc is a useful function that belongs to pandas library, we are using it because data is an object
of pandas. iloc allows us to select specific rows or columns from the dataset
The pattern of using iloc is like this : iloc[specific to rows, specific to columns]
If I write data.iloc[:,0:4] it means that I want to select all the rows(: = all), and which columns ? 
The 0:4 part means : Column 0, column 1, column 2, column 3
'''
X=data.iloc[:,0:4]
print(X) #prints all the rows but just column 0,1,2,3 (not the target column)
# Okay but in our template example, we had a numpy object (something related to numpy library)
'''
In pandas library, there is an attribute that allows us to convert our pandas object into a numpy object
This attribute is values
'''
X=data.iloc[:,0:4].values #converts our pandas dataframe into a numpy object

# Let's store the column target into y variable (remember that "last" in python means -1 in term of index)
y=data.iloc[:,-1].values #again we convert it into numpy object

# Now focus on target, how are the target values
'''
Let's check what are the differents values in target variable with unique(...) function from numpy library 
'''
print(numpy.unique(y)) # prints this ['Iris-setosa' 'Iris-versicolor' 'Iris-virginica']
# So we are in a case where our labels are not integers, 
#  that is why we are going to convert them into integers

# Do the following if label is of type (str)
# begin conversion
# The following code will loop over each element of target variable, and replace "Iris-setosa" by 0,
# "Iris-versicolor" by 1 and "Iris-virginica" by 2
for i in range(len(y)):
    if y[i] == "Iris-setosa":
        y[i] = 0
    elif y[i] == "Iris-versicolor":
        y[i] = 1
    elif y[i] == "Iris-virginica":
        y[i] = 2
# end conversion
# Let's check if conversion worked 
# begin check
print(y)
print(numpy.unique(y)) #prints [0 1 2] so it worked
# end check


# visualizing data
# first we create a object called colors 
colors = numpy.array([x for x in "bgrcmyk"]) 
print(colors) #it prints this : ['b' 'g' 'r' 'c' 'm' 'y' 'k']
# maybe b for blue, g for green, r for red, c for cyan, m for magenta, y for yellow, k for black

# How many features do we have ?
'''
I repeat that X is now a numpy object, remember that we converted it line 61 with attribute values
So it means that we can use functions from numpy on X
A useful function of numpy is shape
'''
print(X.shape) #prints (150, 4) that means X contains 150 rows and 4 columns
# Like we selected only the features columns into X we can say that there is 4 features
# So in plt.scatter, like we are plotting in 2d, we have to specify 2 dimensions to plot data on those
# 4 features means that available indexes are 0,1,2,3. i.e. if we chose to plot our data on dimensions 0 and 2 :
plt.scatter(X[:, 0], X[:, 2], color=colors[list(y)].tolist(), s=10) 
plt.show(); #https://github.com/jupyter/notebook/issues/3523
````

### Why colors are green and black ?
The object *colors* that we created line 30 gives something like this if we print it :
```
['b' 'g' 'r' 'c' 'm' 'y' 'k']
```
Our label values are stored in *y* and with unique function from numpy (line 27) it printed this :
```
[-1  1]
```
Those will be indexes,
So what happens if I access to an element of colors by index 1 :
```
print(colors[1])
```
It will displays this : 
```
g
```
*g* like green.

In fact, the element at index 1 in colors is 'g'.
With the same way, using -1 as index, will take the last element of colors which is *k* (for black)

So what is happening is, all observations that have 1 as label will be plotted in green, and all observations that have -1 as label will be plotted as black

## k nearest neighbour algorithm

### a) Use knn function from sklearn library with k=1
Documentation page : https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

Example template is the following : 
```
X = [[0], [1], [2], [3]]
y = [0, 0, 1, 1]
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y) 
KNeighborsClassifier(...)
print(neigh.predict([[1.1]]))
[0]
print(neigh.predict_proba([[0.9]]))
[[0.66666667 0.33333333]]
```
Let's analyze documentation and this example and keep the useful lines :
- The two first lines refer to the creation of a little dataset of 4 points (1 feature) and 4 labels (one for each point). **We already have X_train, X_test, y_train, y_test, so we don't need those lines**
- Third line is the import of the Knn classifier so we will keep this line
- Line #4 is initialization of knn we keep it but we will modify the value of the parameter to 1
- Line #5 fit function of knn means training (so we will give it our training sets)
- predict function predicts the class labels for the provided data (means we will give it our test set) and returns a list of class labels for each data sample.
- score function returns the mean accuracy on the given test data and labels.


````python
# Import knn
from sklearn.neighbors import KNeighborsClassifier
# Initialize knn with k=1
neigh = KNeighborsClassifier(n_neighbors=1)
# Train knn with our training set
neigh.fit(X_train, y_train) 
# Predict labels for test set with our trained classifier
predictions = neigh.predict(X_test)
print(predictions)
# Evaluate precision of model's predictions by ourself
n_errors = 0
for i in range(len(predictions)):
    if predictions[i] != y_test[i]:
        n_errors += 1
print('Error rate = {}'.format(n_errors/len(predictions)))
print('Precision = {}'.format(1-(n_errors/len(predictions))))
# Evaluate precision of model's predictions using score function
print('Precision = {}'.format(neigh.score(X_test,y_test)))
````

Accuracy (ACC) = 
(Σ True positive + Σ True negative)/Σ Total population

````python
accuracy = (tp+tn)/len(predictions)
print(accuracy)
````

 Recall = Σ True positive / Σ Condition positive

````python
recall = tp/(tp+fn)
print(round(recall,2))
````

````python
recall_neg = tn/(tn+fp)
print(round(recall_neg,2))
````

 Precision = Σ True positive / Σ Predicted condition positive

````python
precision = tp/(fp+tp)
print(round(precision,2))
````

````python
precision_neg = tn/(fn+tn)
print(round(precision_neg,2))
````

F1 score = 2 · Precision · Recall / (Precision + Recall)

````python
f1_score = 2*precision*recall/(precision+recall)
print(round(f1_score,2))
````

````python
f1_score_neg = 2*precision_neg*recall_neg/(precision_neg+recall_neg)
print(round(f1_score_neg,2))
````
